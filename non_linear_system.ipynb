```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

$$
\begin{cases}
\dot{x}= -ay+ bx\sin(x^2+y^2)\\\\
\dot{y}= ax+ by\sin(x^2+y^2)
\end{cases}
\$$

```python
# define the system of differential equations
def system(f,t,a,b):
    dxdt=-a*f[1]+b*f[0]*np.sin(f[0]**2+f[1]**2)
    dydt=a*f[0]+b*f[1]*np.sin(f[0]**2+f[1]**2)
    dfdt=[dxdt,dydt]
    return dfdt

f0 = [np.pi, 0.0]# define initial conditions

# define constants
a = 0.25
b = 5.0

X,Y=np.meshgrid(np.linspace(0,2*np.pi),np.linspace(0,2*np.pi)) # create meshgrid
t = np.linspace(0, 2*np.pi, 50)# define dependent variable
```

```python
# calculate and visualize the solution
sol = odeint(system, f0, t, args=(a, b)) 
plt.plot(t,sol[:,0],'blue',label='x(t)')
plt.plot(t,sol[:,1],'orange',label='y(t)')
plt.legend(loc='best')
plt.xlabel('t')
```

```python
# phase portrait
dxdt,dydt=system((X,Y),t,a,b)
plt.streamplot(X,Y,dxdt,dydt)
```
